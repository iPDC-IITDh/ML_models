```python
import csv
import statistics as st
import random as rd

fp = open('dummy.csv', 'r')
data = csv.reader(fp)
data = list(data)
fp.close()
time_stamp = []
phasor1x = []
phasor1y = []
phasor2x = []
phasor2y = []
phasor3x = []
phasor3y = []
freq = []
dfreq = []
analog1 = []
analog2 = []
for i in data:
    time_stamp.append([i[0]])
    phasor1x.append(float(i[1]))
    phasor1y.append(float(i[2]))
    phasor2x.append(float(i[3]))
    phasor2y.append(float(i[4]))
    phasor3x.append(float(i[5]))
    phasor3y.append(float(i[6]))
    freq.append(float(i[7]))
    dfreq.append(float(i[8]))
    analog1.append(float(i[9]))
    analog2.append(float(i[10]))

mean_phasor1x = st.mean(phasor1x)
stdv_phasor1x = st.stdev(phasor1x)

mean_phasor1y = st.mean(phasor1y)
stdv_phasor1y = st.stdev(phasor1y)

mean_phasor2x = st.mean(phasor2x)
stdv_phasor2x = st.stdev(phasor2x)
mean_phasor2y = st.mean(phasor2y)
stdv_phasor2y = st.stdev(phasor2y)

mean_phasor3x = st.mean(phasor3x)
stdv_phasor3x = st.stdev(phasor3x)
mean_phasor3y = st.mean(phasor3y)
stdv_phasor3y = st.stdev(phasor3y)

mean_freq = st.mean(freq)
stdv_freq = st.stdev(freq)

mean_dfreq = st.mean(dfreq)
stdv_dfreq = st.stdev(dfreq)

mean_analog1 = st.mean(analog1)
stdv_analog1 = st.stdev(analog1)

mean_analog2 = st.mean(analog2)
stdv_analog2 = st.stdev(analog2)

false_data = []
for i in range(len(data)):
    false_data.append(['timeStamp',
                       str(mean_phasor1x-rd.choices([4*stdv_phasor1x,-4*stdv_phasor1x])[0]),
                       str(mean_phasor1y-rd.choices([4*stdv_phasor1y,-4*stdv_phasor1y])[0]),
                       str(mean_phasor2x-rd.choices([4*stdv_phasor2x,-4*stdv_phasor2x])[0]),
                       str(mean_phasor2y-rd.choices([4*stdv_phasor2y,-4*stdv_phasor2y])[0]),
                       str(mean_phasor3x-rd.choices([4*stdv_phasor3x,-4*stdv_phasor3x])[0]),
                       str(mean_phasor3y-rd.choices([4*stdv_phasor3y,-4*stdv_phasor3y])[0]),
                       str(mean_freq-rd.choices([4*stdv_freq,-4*stdv_freq])[0]),
                       str(mean_dfreq-rd.choices([4*stdv_dfreq,-4*stdv_dfreq])[0]),
                       str(mean_analog1-rd.choices([4*stdv_analog1,-4*stdv_analog1])[0]),
                       str(mean_analog2-rd.choices([4*stdv_analog2,-4*stdv_analog2])[0]),
                       '1'])

for i in range(len(data)):
    data[i].append('0')

generated_data = data + false_data

with open('generate_false_data.csv','w') as f:
    writer = csv.writer(f)
    writer.writerows(generated_data)
```
